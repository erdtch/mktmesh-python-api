```python
import pandas as pd
import numpy as np

df = pd.read_csv('tick.csv', names=['time', 'close', 'volume'])
df.tail(5)
```

```python
df['time'] = pd.to_datetime(df['time'])
df = df.set_index('time')
df.tail(20)
```

```python
ohlc = df['close'].resample('1Min').ohlc().fillna(method='bfill')
ohlc.tail(10)
```

```python
volume = df['volume'].resample('1Min').sum().fillna(0)
volume.tail(10)
```

```python
ohlcv = pd.concat([ohlc, volume], axis=1)
ohlcv.tail(10)
```

```python
ohlcv['bar_index'] = np.arange(len(ohlcv))
ohlcv.tail(10)
```
