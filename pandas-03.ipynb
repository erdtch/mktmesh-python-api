```python
import pandas as pd
from io import StringIO
import urllib
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import ticker as ticker
from mpl_finance import candlestick_ohlc
import matplotlib.dates as mdates
import requests
```

```python
# read url content with key
# TODO change to http://122.8.148.106
url = 'http://122.8.148.106/api/v1/data/candlestick/000/1032887/?symbol=PTT&timeframe=D&startTime=2018-03-01&endTime=2018-08-18'
headers = {'X-API-KEY': '907E0507067041318FBD5248250E4EF1'}
response = requests.get(url, headers=headers)
data = response.text

df = pd.read_csv(StringIO(data), names=['time', 'open', 'high', 'low', 'close', 'volume', 'oi'], parse_dates=True)
df
```

```python
df['time'] = pd.to_datetime(df['time'])
df["time"] = df["time"].apply(mdates.date2num)
df
```

```python
f1, ax = plt.subplots(figsize = (10,5))

# plot the candlesticks
candlestick_ohlc(ax, df.values, width=.6, colorup='green', colordown='red')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
ax.xaxis.set_major_locator(ticker.MaxNLocator(6))

f1.autofmt_xdate()
f1.tight_layout()

plt.show()
```
